```python
import numpy as np

probs = {'w1': 0.50, 'w2': 0.10, 'w3': 0.20, 'w4': 0.20}
X = {'w1': 0, 'w2': 1, 'w3': 0, 'w4': 1}
Y = {'w1': 0, 'w2': 0, 'w3': 1, 'w4': 1}
omegas = ['w1', 'w2', 'w3', 'w4']

P_Y0 = probs['w1'] + probs['w2']
P_Y1 = probs['w3'] + probs['w4']
P_Y  = {0: P_Y0, 1: P_Y1}

E_X_given_Y0 = (X['w2'] * probs['w2']) / P_Y0
E_X_given_Y1 = (X['w4'] * probs['w4']) / P_Y1
E_X_Y = {w: E_X_given_Y0 if Y[w] == 0 else E_X_given_Y1 for w in omegas}
joint = {(X[w], Y[w]): probs[w] for w in omegas}

print("Part 1: F = P(Omega) = all 16 subsets,", probs)

print("\nPart 2: sigma(Y) = { {}, {w1,w2}, {w3,w4}, Omega }")

print("\nPart 3: E[X|Y] must be constant on atoms {w1,w2} and {w3,w4}")

print(f"\nPart 4: E[X|Y=0]={E_X_given_Y0:.4f}, E[X|Y=1]={E_X_given_Y1:.4f}")

print("\nPart 5:")
for name, group in {'{w1,w2}': ['w1','w2'], '{w3,w4}': ['w3','w4'], 'Omega': omegas}.items():
    lhs = sum(E_X_Y[w] * probs[w] for w in group)
    rhs = sum(X[w] * probs[w] for w in group)
    print(f"  A={name}: LHS={lhs:.4f}, RHS={rhs:.4f} => {'CHECK' if abs(lhs-rhs)<1e-9 else 'FAIL'}")

print("\nPart 6:")
for y in [0, 1]:
    for x in [0, 1]:
        print(f"  p(X={x}|Y={y}) = {joint.get((x,y),0)/P_Y[y]:.4f}")

print("\nPart 7:")
for y in [0, 1]:
    val = sum(x * (joint.get((x,y),0)/P_Y[y]) for x in [0,1])
    exp = E_X_given_Y0 if y==0 else E_X_given_Y1
    print(f"  y={y}: {val:.4f} == {exp:.4f} => {'CHECK' if abs(val-exp)<1e-9 else 'FAIL'}")

MSE = sum((X[w] - E_X_Y[w])**2 * probs[w] for w in omegas)
print(f"\nPart 8: MSE = {MSE:.4f} = 11/60")

E_X = sum(X[w] * probs[w] for w in omegas)
MSE_global = sum((X[w] - E_X)**2 * probs[w] for w in omegas)
Var_EXY = MSE_global - MSE
print(f"\nPart 9: E[X]={E_X:.4f}, MSE_global={MSE_global:.4f}")
print(f"  Var(X) = E[Var(X|Y)] + Var(E[X|Y]) => {MSE_global:.4f} = {MSE:.4f} + {Var_EXY:.4f}")

def entropy(p_list):
    return -sum(p * np.log2(p) for p in p_list if p > 0)

HX   = entropy([sum(probs[w] for w in omegas if X[w]==x) for x in [0,1]])
HX_Y = sum(P_Y[y] * entropy([joint.get((x,y),0)/P_Y[y] for x in [0,1]]) for y in [0,1])
IXY  = HX - HX_Y

print(f"\nPart 10:")
print(f"  H(X)   = {HX:.4f} bits")
print(f"  H(X|Y) = {HX_Y:.4f} bits")
print(f"  I(X;Y) = {IXY:.4f} bits")
print(f"  Uncertainty removed = {IXY/HX*100:.2f}%")
```

# Q2) Conditional Density of the Multivariate Gaussian

Let $\mathbf{Z}$ be the concatenated random matrix
$\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$.


The random variable $\mathbf{Z}$ is said to have a **joint multivariate normal distribution** if every linear combination of its components follows a univariate normal distribution.

For $\mathbf{Z}$ to possess a **joint density function** (meaning its distribution is absolutely continuous with respect to the Lebesgue measure on $\mathbb{R}^{n+m}$), its covariance matrix must be non-singular (positive definite).


The joint density function $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ for $\mathbf{z} \in \mathbb{R}^{n+m}$ is given by:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right).$$

Here:

* **Mean ($\boldsymbol{\mu}$):** $\boldsymbol{\mu} = E[\mathbf{Z}] = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix} \in \mathbb{R}^{n+m}$, where $\boldsymbol{\mu}_X \in \mathbb{R}^n$ and $\boldsymbol{\mu}_Y \in \mathbb{R}^m$.
* **Covariance Matrix ($\boldsymbol{\Sigma}$):** A symmetric, positive definite matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{(n+m) \times (n+m)}$ partitioned as:
$$\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}$$
    * $\boldsymbol{\Sigma}_{XX} = \text{Var}(\mathbf{X})$ (size $n \times n$)
    * $\boldsymbol{\Sigma}_{YY} = \text{Var}(\mathbf{Y})$ (size $m \times m$)
    * $\boldsymbol{\Sigma}_{XY} = \text{Cov}(\mathbf{X}, \mathbf{Y}) = \boldsymbol{\Sigma}_{YX}^T$ (size $n \times m$)
* **Determinant ($|\boldsymbol{\Sigma}|$):** The normalization factor ensuring the total integral is 1.

1. Show that the conditional distribution is given by $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right),$$
where
* $\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$

Q2) CONDITIONAL DENSITY OF THE MULTIVARIATE GAUSSIAN
=====================================================

GOAL
  Show that f_{X|Y}(x|y) is multivariate normal:
  f_{X|Y}(x|y) ~ N(mu*, Sigma_{X|Y})

  with:
    Sigma_{X|Y} = Sigma_XX - Sigma_XY Sigma_YY^{-1} Sigma_YX
    mu*         = mu_X + Sigma_XY Sigma_YY^{-1} (y - mu_Y)


STEP 1: Definition of conditional density

  f_{X|Y}(x|y) = f_{X,Y}(x,y) / f_Y(y)


STEP 2: Marginal of Y

  Since Y is a sub-vector of a jointly normal Z:
    Y ~ N(mu_Y, Sigma_YY)

  f_Y(y) = 1 / sqrt((2pi)^m |Sigma_YY|)
           * exp( -1/2 (y-mu_Y)^T Sigma_YY^{-1} (y-mu_Y) )


STEP 3: Block inverse of Sigma

  Let a = x - mu_X,  b = y - mu_Y.

  Using the Schur complement S = Sigma_{X|Y}, block
  inversion of Sigma gives the quadratic form:

    (z-mu)^T Sigma^{-1} (z-mu)
      = (a - Sigma_XY Sigma_YY^{-1} b)^T S^{-1}
        (a - Sigma_XY Sigma_YY^{-1} b)
        + b^T Sigma_YY^{-1} b


STEP 4: Complete the square (define mu*)

  Set:
    mu*     = mu_X + Sigma_XY Sigma_YY^{-1} (y - mu_Y)
    x - mu* = a - Sigma_XY Sigma_YY^{-1} b

  Then the quadratic form splits cleanly:

    (z-mu)^T Sigma^{-1} (z-mu)
      = (x-mu*)^T Sigma_{X|Y}^{-1} (x-mu*)
        + (y-mu_Y)^T Sigma_YY^{-1} (y-mu_Y)


STEP 5: Block determinant

  By the Schur complement determinant formula:
    |Sigma| = |Sigma_YY| * |Sigma_{X|Y}|


STEP 6: Factor the joint density

  Substituting Steps 4 and 5 into f_{X,Y}(x,y):

    f_{X,Y}(x,y)
      = [ 1/sqrt((2pi)^n |Sigma_{X|Y}|)
          * exp(-1/2 (x-mu*)^T Sigma_{X|Y}^{-1} (x-mu*)) ]
        x
        [ 1/sqrt((2pi)^m |Sigma_YY|)
          * exp(-1/2 (y-mu_Y)^T Sigma_YY^{-1} (y-mu_Y)) ]

      = f_{X|Y}(x|y) * f_Y(y)


STEP 7: Conclude

  Dividing by f_Y(y):

    f_{X|Y}(x|y)
      = 1 / sqrt((2pi)^n |Sigma_{X|Y}|)
        * exp( -1/2 (x-mu*)^T Sigma_{X|Y}^{-1} (x-mu*) )

  This is exactly N(mu*, Sigma_{X|Y}).   QED

  where:
    Sigma_{X|Y} = Sigma_XX - Sigma_XY Sigma_YY^{-1} Sigma_YX
                  (conditional covariance)
    mu*         = mu_X + Sigma_XY Sigma_YY^{-1} (y - mu_Y)
                  (conditional mean)

```python
import numpy as np

# Setup: example with n=2, m=2
np.random.seed(42)

mu_X = np.array([1.0, 2.0])
mu_Y = np.array([3.0, 4.0])
mu   = np.concatenate([mu_X, mu_Y])

# Full covariance matrix (must be positive definite)
Sigma = np.array([
    [4.0, 2.0, 1.0, 0.5],
    [2.0, 3.0, 0.5, 1.0],
    [1.0, 0.5, 2.0, 0.8],
    [0.5, 1.0, 0.8, 1.5]
])

n = len(mu_X)
m = len(mu_Y)

# Partition covariance matrix
Sigma_XX = Sigma[:n, :n]
Sigma_XY = Sigma[:n, n:]
Sigma_YX = Sigma[n:, :n]
Sigma_YY = Sigma[n:, n:]

# Conditional covariance and mean
Sigma_YY_inv  = np.linalg.inv(Sigma_YY)
Sigma_X_given_Y = Sigma_XX - Sigma_XY @ Sigma_YY_inv @ Sigma_YX

y        = np.array([2.0, 3.0])
mu_star  = mu_X + Sigma_XY @ Sigma_YY_inv @ (y - mu_Y)

print("Part 1: Partitioned covariance blocks")
print(f"  Sigma_XX:\n{Sigma_XX}")
print(f"  Sigma_XY:\n{Sigma_XY}")
print(f"  Sigma_YY:\n{Sigma_YY}")

print(f"\nPart 2: Conditional covariance Sigma_{{X|Y}}")
print(f"  Sigma_{{X|Y}} = Sigma_XX - Sigma_XY @ Sigma_YY^-1 @ Sigma_YX")
print(f"\n{Sigma_X_given_Y}")

print(f"\nPart 3: Conditional mean mu*")
print(f"  mu* = mu_X + Sigma_XY @ Sigma_YY^-1 @ (y - mu_Y)")
print(f"  mu* = {mu_star}")

# Verify factorization: |Sigma| = |Sigma_YY| * |Sigma_{X|Y}|
det_Sigma    = np.linalg.det(Sigma)
det_YY       = np.linalg.det(Sigma_YY)
det_X_given_Y = np.linalg.det(Sigma_X_given_Y)

print(f"\nPart 4: Determinant factorization |Sigma| = |Sigma_YY| * |Sigma_{{X|Y}}|")
print(f"  |Sigma|              = {det_Sigma:.6f}")
print(f"  |Sigma_YY|*|Sig_X|Y| = {det_YY * det_X_given_Y:.6f}")
print(f"  CHECK: {np.isclose(det_Sigma, det_YY * det_X_given_Y)}")

# Verify quadratic form decomposition
x = np.array([1.5, 2.5])
z = np.concatenate([x, y])

a = x - mu_X
b = y - mu_Y
Sigma_inv = np.linalg.inv(Sigma)

lhs = (z - mu) @ Sigma_inv @ (z - mu)

Sigma_X_given_Y_inv = np.linalg.inv(Sigma_X_given_Y)
term1 = (x - mu_star) @ Sigma_X_given_Y_inv @ (x - mu_star)
term2 = b @ Sigma_YY_inv @ b
rhs = term1 + term2

print(f"\nPart 5: Quadratic form decomposition")
print(f"  (z-mu)^T Sigma^-1 (z-mu)                        = {lhs:.6f}")
print(f"  (x-mu*)^T Sig_X|Y^-1 (x-mu*) + (y-mu_Y)^T ..  = {rhs:.6f}")
print(f"  CHECK: {np.isclose(lhs, rhs)}")

# Conditional density at x given y
def mvn_density(x, mu, Sigma):
    k   = len(x)
    det = np.linalg.det(Sigma)
    inv = np.linalg.inv(Sigma)
    c   = 1.0 / np.sqrt((2*np.pi)**k * det)
    q   = (x - mu) @ inv @ (x - mu)
    return c * np.exp(-0.5 * q)

f_X_given_Y = mvn_density(x, mu_star, Sigma_X_given_Y)

# Verify: f_{X|Y}(x|y) = f_{X,Y}(x,y) / f_Y(y)
f_joint = mvn_density(z, mu, Sigma)
f_Y     = mvn_density(y, mu_Y, Sigma_YY)

print(f"\nPart 6: Verify f_{{X|Y}}(x|y) = f_{{X,Y}}(x,y) / f_Y(y)")
print(f"  f_{{X|Y}} direct         = {f_X_given_Y:.8f}")
print(f"  f_{{X,Y}} / f_Y          = {f_joint/f_Y:.8f}")
print(f"  CHECK: {np.isclose(f_X_given_Y, f_joint/f_Y)}")

print(f"\nConclusion:")
print(f"  X|Y=y ~ N(mu*, Sigma_{{X|Y}})")
print(f"  mu*          = {mu_star}")
print(f"  Sigma_{{X|Y}} =\n{Sigma_X_given_Y}")
```

Q3) CONTINUOUS FAULT MONITORING OF A SPINDLE MOTOR


SETUP
  X ~ N(0, 1)        hidden temperature (standardized)
  Q ~ N(0, 4)        mechanical load, independent of X
  Y = X + Q          measured spindle power


PART 1: Joint Distribution

(X, Y) is jointly (bivariate) normal, since X and Q are independent
normals and Y = X + Q is a linear combination of them.

  E[Y]   = E[X] + E[Q] = 0 + 0 = 0
  Var(Y) = Var(X) + Var(Q) = 1 + 4 = 5      (independence)
  => Y ~ N(0, 5)   CHECK

  Cov(X, Y) = Cov(X, X + Q) = Cov(X,X) + Cov(X,Q)
            = Var(X) + 0 = 1

  (X, Y) ~ N( [0, 0],  [[1, 1], [1, 5]] )


PART 2: Information Geometry - Why X is NOT sigma(Y)-measurable

X is not a deterministic function of Y, so it cannot be sigma(Y)-
measurable. The same Y can arise from infinitely many (X, Q) pairs.

Practical meaning: if the meter shows Y = 5, this could be either a
HIGH temperature X with LOW torque Q, or a LOW temperature X with HIGH
torque Q. The single measurement confounds the thermal effect with the
mechanical effect, so the engineer cannot determine X with certainty.


PART 3: Conditional Expectation (Best Estimate)

  E[X|Y=y] = mu_X + [Cov(X,Y)/Var(Y)] (y - mu_Y)
           = 0 + (1/5)(y - 0)
           = y/5

  => E[X|Y] = Y/5

Signal-to-total-variance ratio:
  Cov(X,Y)/Var(Y) = Var(X)/Var(Y) = 1/5 = 0.20

Only 20% of the variance in the power reading comes from the true
temperature signal; the rest is mechanical noise.


PART 4: Verification of the Residual

  X_hat = E[X|Y] = Y/5
  Z     = X - X_hat = X - Y/5

E[Z] = E[X] - E[Y]/5 = 0 - 0 = 0   CHECK

Orthogonality (Z uncorrelated with Y):
  Cov(Z, Y) = Cov(X - Y/5, Y)
            = Cov(X,Y) - (1/5)Var(Y)
            = 1 - (1/5)(5)
            = 0   CHECK


PART 5: Differential Entropy

Formula: h = (1/2) log2( 2*pi*e * variance )

h(X), with Var(X) = 1:
  h(X) = (1/2) log2(2*pi*e * 1) = 2.0471 bits

Conditional variance:
  sigma^2_{X|Y} = sigma^2_X - Cov(X,Y)^2 / Var(Y)
                = 1 - (1)^2 / 5
                = 4/5 = 0.80

h(X|Y), with variance 0.80:
  h(X|Y) = (1/2) log2(2*pi*e * 0.80) = 1.8861 bits


PART 6: Mutual Information

  I(X;Y) = h(X) - h(X|Y)
         = (1/2) log2( 1 / 0.80 )
         = (1/2) log2(1.25)
         = 0.1610 bits

Equivalent formula using correlation:
  rho^2  = Cov(X,Y)^2 / (Var(X)Var(Y)) = 1/(1*5) = 0.20
  I(X;Y) = -(1/2) log2(1 - rho^2) = -(1/2) log2(0.80) = 0.1610 bits

Percentage of uncertainty removed:
  Variance reduction = rho^2 = 20%
  (Observing the power meter removes 20% of the variance in X.)

If Q had variance 100 instead of 4:
  Var(Y) = 1 + 100 = 101
  rho^2  = 1/101 ≈ 0.0099  (≈ 0.99%)
  I(X;Y) = -(1/2) log2(1 - 1/101) ≈ 0.0072 bits

Almost no information is gained: the large mechanical noise drowns out
the temperature signal, so the meter becomes nearly useless for fault
detection.


PART 7: Diagnostic Interpretation

The manager's suggestion (use Y itself as the estimate of X) is wrong.
Since Y = X + Q, the raw reading contains the mechanical noise Q, so it
is a BIASED estimator of X - it systematically overstates the magnitude
of temperature deviations.

The optimal estimator E[X|Y] = Y/5 SHRINKS the measurement toward the
prior mean (zero) by the factor 1/5 = signal-to-total-variance ratio.
This shrinkage is correct because only 1/5 of any observed Y is due to
the true temperature signal; the rest is noise. The conditional
expectation optimally balances the prior knowledge (X centered at 0)
against the noisy evidence Y, minimizing mean square error.

```python
import numpy as np

# Setup
var_X = 1.0
var_Q = 4.0
mu_X = mu_Q = 0.0

# Part 1: Joint Distribution
mu_Y    = mu_X + mu_Q
var_Y   = var_X + var_Q
cov_XY  = var_X  # Cov(X, X+Q) = Var(X)

print("Part 1: Joint Distribution")
print(f"  E[Y]   = {mu_Y}")
print(f"  Var(Y) = Var(X)+Var(Q) = {var_X}+{var_Q} = {var_Y}  => Y ~ N(0,5)")
print(f"  Cov(X,Y) = Var(X) = {cov_XY}")
print(f"  (X,Y) ~ N([0,0], [[{var_X},{cov_XY}],[{cov_XY},{var_Y}]])")

# Part 2
print("\nPart 2: X is NOT sigma(Y)-measurable")
print("  X is not a deterministic function of Y; same Y from many (X,Q).")
print("  Y=5 could be high X/low Q or low X/high Q -> confounded.")

# Part 3: Conditional Expectation
beta = cov_XY / var_Y  # signal-to-total-variance ratio
print("\nPart 3: Conditional Expectation")
print(f"  E[X|Y] = mu_X + (Cov(X,Y)/Var(Y))(Y-mu_Y) = (1/5)Y")
print(f"  Signal-to-total-variance ratio = Cov(X,Y)/Var(Y) = {beta:.4f}")

# Part 4: Residual verification (numerical)
np.random.seed(0)
N = 2_000_000
X = np.random.normal(0, np.sqrt(var_X), N)
Q = np.random.normal(0, np.sqrt(var_Q), N)
Y = X + Q
X_hat = beta * Y
Z = X - X_hat

print("\nPart 4: Verification of the Residual")
print(f"  E[Z]      ≈ {np.mean(Z):.5f}  (theory 0)")
print(f"  Cov(Z,Y)  ≈ {np.cov(Z, Y)[0,1]:.5f}  (theory 0, orthogonality)")

# Part 5: Differential Entropy
def diff_entropy(var):
    return 0.5 * np.log2(2 * np.pi * np.e * var)

var_X_given_Y = var_X - cov_XY**2 / var_Y
hX   = diff_entropy(var_X)
hX_Y = diff_entropy(var_X_given_Y)

print("\nPart 5: Differential Entropy")
print(f"  sigma^2_{{X|Y}} = {var_X} - {cov_XY}^2/{var_Y} = {var_X_given_Y:.4f}")
print(f"  h(X)   = {hX:.4f} bits")
print(f"  h(X|Y) = {hX_Y:.4f} bits")

# Part 6: Mutual Information
IXY = hX - hX_Y
rho2 = cov_XY**2 / (var_X * var_Y)
IXY_alt = -0.5 * np.log2(1 - rho2)

print("\nPart 6: Mutual Information")
print(f"  I(X;Y) = h(X)-h(X|Y) = {IXY:.4f} bits")
print(f"  I(X;Y) = -0.5*log2(1-rho^2) = {IXY_alt:.4f} bits  (CHECK)")
print(f"  Variance reduction (rho^2) = {rho2*100:.2f}%")

# If Q variance = 100
var_Q2  = 100.0
var_Y2  = var_X + var_Q2
rho2_2  = var_X**2 / (var_X * var_Y2)
IXY2    = -0.5 * np.log2(1 - rho2_2)
print(f"\n  If Var(Q)=100: Var(Y)={var_Y2}, rho^2={rho2_2:.4f} ({rho2_2*100:.2f}%)")
print(f"  I(X;Y) = {IXY2:.4f} bits  (signal drowned out by noise)")

# Part 7
print("\nPart 7: Diagnostic Interpretation")
print("  Y = X+Q contains noise Q, so Y is a BIASED estimator of X.")
print(f"  Optimal E[X|Y] = (1/5)Y shrinks measurement toward 0 by factor {beta:.2f}.")
print("  Only 1/5 of Y is true signal; shrinkage minimizes mean square error.")
```
