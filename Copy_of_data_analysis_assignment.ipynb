### Question 1: The Total Variance Illusion (PCA Subplots 1 & 2 vs. FA Subplots 1 & 2)

* **Context:** In your PCA dashboard, `PC 1` captures roughly $45\%$ of the total system variance, and `PC 2` captures over $30\%$, indicating a highly structured, multi-component system. However, looking at the Factor Analysis dashboard, the **Variance Partitioning Stacked Bar** reveals that `Sensor_4` has a **Uniqueness value ($\varphi^2$) exceeding $98\%$**.
* **Core Problem:** 1. What does a Uniqueness value close to $100\%$ tell you about the physical nature of the data coming from `Sensor_4`?
2. Because PCA maximizes total global variance without distinguishing between types of variance, how does the massive localized noise of `Sensor_4` "trick" the PCA engine into inflating its top eigenvalues?
3. Why is this a major risk if an engineer relies *only* on PCA to build an automated plant anomaly detection framework?



````python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA, FactorAnalysis
from sklearn.preprocessing import StandardScaler

# 1. Generate Synthetic Plant Data
np.random.seed(42)
n_samples = 1000

true_process = np.sin(np.linspace(0, 20, n_samples))
sensor_1 = true_process + np.random.normal(0, 0.5, n_samples)
sensor_2 = true_process * 1.5 + np.random.normal(0, 0.5, n_samples)
sensor_3 = true_process * 0.8 + np.random.normal(0, 0.5, n_samples)
sensor_4 = np.random.normal(0, 10.0, n_samples)  # Pure noise, massive variance

df = pd.DataFrame({
    'Sensor_1': sensor_1, 'Sensor_2': sensor_2,
    'Sensor_3': sensor_3, 'Sensor_4': sensor_4
})

print("--- The Total Variance Illusion ---\n")

# 2. Standardise first (required for FA uniqueness to be mathematically valid)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)  # Each sensor: mean=0, std=1

# 3. Run PCA on standardised data (correlation-matrix PCA)
pca = PCA(n_components=2)
pca.fit(X_scaled)

print("PCA RESULTS:")
print(f"  Variance explained by PC1: {pca.explained_variance_ratio_[0]:.2%}")
print(f"  Variance explained by PC2: {pca.explained_variance_ratio_[1]:.2%}")
print(f"  PC1 Loadings (note how Sensor_4 dominates):")
print(pd.Series(pca.components_[0], index=df.columns).round(3).to_string())

# 4. Run Factor Analysis on standardised data
# On standardised data, noise_variance_ IS ψ² directly (each variable has total variance = 1)
# No manual division needed — unlike operating on raw unstandardised data
fa = FactorAnalysis(n_components=1, random_state=42)
fa.fit(X_scaled)

uniqueness = fa.noise_variance_  # ψ² per sensor, valid range [0, 1]

print("\nFACTOR ANALYSIS RESULTS:")
print("Uniqueness ψ² — variance NOT shared with the latent plant process:\n")
for sensor, val in zip(df.columns, uniqueness):
    bar = '█' * int(val * 20)
    flag = " ← MOSTLY NOISE (faulty/isolated sensor)" if val > 0.90 else ""
    print(f"  {sensor}: {val:.2%}  {bar}{flag}")

print("\nINTERPRETATION:")
print("  Sensors 1-3: low ψ² → most variance is SHARED with the true process.")
print("  Sensor_4:    ψ² ≈ 98% → almost NO shared information.")
print("  PCA cannot see this distinction — it inflates PC1/PC2 using")
print("  Sensor_4's noise, making the system look more 'explained' than it is.")
print("  This causes false alarms and missed detections in anomaly frameworks.")
````

In Factor Analysis, every sensor's total variance is partitioned into two components: Communality, which is the portion explained by shared latent factors (the real plant process), and Uniqueness (ψ²), which is the portion that is entirely sensor-specific and not shared with anything else in the system.
A Uniqueness value approaching 100% means that virtually none of Sensor_4's variance is connected to any underlying plant process. The sensor is fluctuating independently of everything else. Physically, this indicates the sensor is either faulty or poorly calibrated, picking up electromagnetic interference or instrument noise, or genuinely measuring a physical variable that is completely decoupled from the rest of the process. In any case, Sensor_4 carries essentially zero process information — it is a noise source masquerading as a measurement.
PCA finds directions of maximum total variance in the data without distinguishing between variance that comes from real process dynamics versus variance that comes from sensor noise. It treats all variance as equally meaningful. Sensor_4 has a standard deviation of 10, giving it a variance of approximately 100, which is roughly 100 times larger than the other sensors. When PCA examines the covariance matrix, this enormous diagonal entry for Sensor_4 dominates. PCA's eigenvectors are forced to rotate toward Sensor_4's noise direction to explain this large variance. The result is that PC1 ends up carrying a heavy loading on Sensor_4, and the explained variance percentage reported for PC1 and PC2 appears impressively large, but a significant portion of that explained variance is just captured noise and not captured process structure. The illusion is that the scree plot looks like the system is well-described by two components, when in reality the top components are contaminated by an uninformative sensor.
Relying only on PCA creates two simultaneous failure modes, both of which are unacceptable in a safety-critical plant environment. The first is false alarms. When Sensor_4 produces a random noise burst, PCA's anomaly score spikes above the threshold, the system raises an alarm, engineers investigate, and find nothing wrong. Repeated false alarms cause alarm fatigue, where operators begin ignoring alerts, meaning that when a real fault eventually occurs it also gets dismissed. The second is missed alarms. Because Sensor_4's noise inflates the baseline variance captured by the principal components, the anomaly detection threshold is set artificially high. A genuine process fault such as an abnormal temperature surge in Sensors 1 and 2 produces a shift in the shared latent factor, but this real signal is drowned out by the noise floor that Sensor_4 has established, and the real event goes undetected. Factor Analysis avoids both failures because it explicitly separates shared variance from unique noise, monitoring only the communal signal space and ignoring sensor-specific noise, making it simultaneously less sensitive to instrument noise and more sensitive to genuine process deviations.

### Question 2: Decoupling Structural Loading via Rotation (FA Subplot 1 vs. PCA Eigenvectors)

* **Context:** Look at the **Structural Loadings Matrix** heatmap in the top-left of the FA dashboard. You can see that `Sensor_1` and `Sensor_2` load heavily on `Factor 1`, while `Sensor_3` loads exclusively on `Factor 2`.
* **Core Problem:**
1. Traditional PCA forces its principal axes to follow a strict mathematical hierarchy ($\lambda_1 > \lambda_2 > \dots$) where every component tries to sweep up as much mixed variance as possible across all sensors. Explain how the **Varimax rotation** in Factor Analysis relaxes this rule to create "simple structure" (clean, isolated groupings).
2. From a plant operator's standpoint, why is the rotated FA heatmap much easier to troubleshoot during a structural failure than a raw, unrotated PCA eigenvector matrix?





````python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA, FactorAnalysis
from sklearn.preprocessing import StandardScaler

np.random.seed(42)
n_samples = 1000

# Two distinct latent plant processes
process_A = np.sin(np.linspace(0, 20, n_samples))  # e.g. temperature loop
process_B = np.cos(np.linspace(0, 10, n_samples))  # e.g. pressure loop

# Sensor 1 & 2 tied to process A, Sensor 3 tied to process B, Sensor 4 is noise
sensor_1 = 0.90 * process_A + np.random.normal(0, 0.20, n_samples)
sensor_2 = 0.85 * process_A + np.random.normal(0, 0.25, n_samples)
sensor_3 = 0.92 * process_B + np.random.normal(0, 0.20, n_samples)
sensor_4 = np.random.normal(0, 0.30, n_samples)

df = pd.DataFrame({
    'Sensor_1': sensor_1, 'Sensor_2': sensor_2,
    'Sensor_3': sensor_3, 'Sensor_4': sensor_4
})

print("--- Decoupling Structural Loading via Rotation ---\n")

# Standardise first
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

# 1. PCA — no rotation, follows strict λ1 > λ2 hierarchy
pca = PCA(n_components=2)
pca.fit(X_scaled)

print("PCA RESULTS (no rotation):")
print(f"Variance explained: PC1={pca.explained_variance_ratio_[0]:.2%}, PC2={pca.explained_variance_ratio_[1]:.2%}")
print("PC Loadings (every sensor loads on BOTH components — mixed structure):")
pca_loadings = pd.DataFrame(
    pca.components_.T,
    index=df.columns,
    columns=['PC1', 'PC2']
)
print(pca_loadings.round(3))

# 2. Factor Analysis WITH Varimax rotation
fa = FactorAnalysis(n_components=2, rotation='varimax', random_state=42)
fa.fit(X_scaled)

print("\nFACTOR ANALYSIS RESULTS (with Varimax rotation):")
print("FA Loadings (simple structure — each sensor loads on ONE factor only):")
fa_loadings = pd.DataFrame(
    fa.components_.T,
    index=df.columns,
    columns=['Factor_1', 'Factor_2']
)
print(fa_loadings.round(3))

print("\nINTERPRETATION:")
print("  PCA: All sensors partially load on PC1 AND PC2.")
print("       An alarm on PC1 gives no clear subsystem to investigate.")
print("  FA:  Sensor_1 & Sensor_2 → Factor 1 (temperature loop)")
print("       Sensor_3             → Factor 2 (pressure loop)")
print("       Sensor_4             → ~0 on both (noise/faulty sensor)")
print("  Varimax maximises variance of squared loadings, pushing each")
print("  value toward 0 or ±1 — creating clean, isolated sensor groups.")
print("  Total variance explained is IDENTICAL before and after rotation.")
print("  The rotation only changes interpretability, not information content.")
````

PCA forces its principal components to follow a strict mathematical hierarchy where each component must capture the maximum possible total variance across all sensors simultaneously. Because of this constraint, every sensor ends up with non-trivial loadings on both the first and second components, creating a mixed structure where no single component maps cleanly to a physical subsystem. The eigenvectors point in mathematically optimal directions, but not physically meaningful ones.
Varimax rotation relaxes this hierarchy entirely. After the unrotated factors are extracted, Varimax rotates the factor axes to maximise the variance of the squared loadings within each factor. This mathematically forces each loading toward either zero or one, meaning each sensor ends up loading heavily on one factor and near-zero on all others. This is called simple structure. Crucially, the total variance explained is identical before and after rotation. Varimax only changes where the axes point, not how much information is captured.
In this example, after Varimax rotation, Sensor 1 and Sensor 2 load exclusively on Factor 1, Sensor 3 loads exclusively on Factor 2, and Sensor 4 is near-zero on both factors, correctly flagging it as a noise source. This clean separation directly reflects the two underlying physical processes in the plant.
From a plant operator's standpoint, the rotated FA heatmap is far easier to use during a structural failure because the loading matrix becomes a direct diagnosis guide. With PCA, when an alarm fires, the operator must suspect all sensors simultaneously because every sensor has a meaningful loading on every component. There is no clear mapping from a component anomaly to a specific subsystem, so root cause isolation requires additional statistical analysis every time.
With the rotated FA heatmap, the mapping is immediate. If Factor 1 spikes, only Sensor 1 and Sensor 2 are relevant and the operator investigates the temperature loop. If Factor 2 spikes, only Sensor 3 is relevant and the operator investigates the pressure loop. No statistical expertise is required at alarm time. The physical subsystems map directly to the factors, and Sensor 4's near-zero loadings on both factors flag it for inspection or decommissioning. In a safety-critical plant environment, this interpretability directly translates to faster response times and fewer missed or misdiagnosed faults.

### Question 3: Determining Subspace Truncation ($k$) using $T^2$ and $Q$ Profiles

* **Context:** Look at the right side of the PCA dashboard, showing **Mean Hotelling's $T^2$** and **Mean $Q$ Statistic (SPE)** plotted against the truncation cutoff $k$.
* **Core Problem:**
1. Describe the exact behavior of the Mean $Q$ Statistic curve as the subspace cutoff transitions from $k=1$ to $k=2$, and then from $k=2$ to $k=3$.
2. Why does a sharp drop followed by a distinct "flat elbow" in the $Q$ statistic identify **$k=2$** as the true hidden physical dimensionality of this asset?
3. If you incorrectly choose a truncation cutoff of $k=3$ for your monitoring system, what type of data are you forcing into your "clean" principal subspace?

````python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

np.random.seed(42)
n_samples = 1000

# Two true latent processes (real physical dimensionality = 2)
process_A = np.sin(np.linspace(0, 20, n_samples))
process_B = np.cos(np.linspace(0, 10, n_samples))

# Sensors 1 and 2 driven by process A, Sensor 3 driven by process B
# Sensors 4 and 5 are pure noise — no latent process drives them
sensor_1 = 0.90 * process_A + np.random.normal(0, 0.20, n_samples)
sensor_2 = 0.85 * process_A + np.random.normal(0, 0.25, n_samples)
sensor_3 = 0.92 * process_B + np.random.normal(0, 0.20, n_samples)
sensor_4 = np.random.normal(0, 1.0, n_samples)
sensor_5 = np.random.normal(0, 1.0, n_samples)

df = pd.DataFrame({
    'Sensor_1': sensor_1, 'Sensor_2': sensor_2, 'Sensor_3': sensor_3,
    'Sensor_4': sensor_4, 'Sensor_5': sensor_5
})

print("--- Determining Subspace Truncation using T² and Q Profiles ---\n")

# Standardise
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

# Full PCA (all components)
pca_full = PCA(n_components=5)
pca_full.fit(X_scaled)

print("Variance explained per component:")
for i, ev in enumerate(pca_full.explained_variance_ratio_):
    print(f"  PC{i+1}: {ev:.2%}")

# Compute Mean T² and Mean Q (SPE) for each truncation k = 1 to 5
k_values = list(range(1, 6))
mean_T2  = []
mean_Q   = []

for k in k_values:
    pca_k = PCA(n_components=k)
    scores    = pca_k.fit_transform(X_scaled)          # projection into k-subspace
    X_recon   = pca_k.inverse_transform(scores)        # reconstruction from k PCs

    # T² — variance captured INSIDE the subspace (Hotelling)
    # Mean T² across all samples
    eigenvalues = pca_k.explained_variance_
    T2_scores   = np.sum((scores ** 2) / eigenvalues, axis=1)
    mean_T2.append(np.mean(T2_scores))

    # Q (SPE) — residual variance LEFT OUTSIDE the subspace
    residuals = X_scaled - X_recon
    Q_scores  = np.sum(residuals ** 2, axis=1)
    mean_Q.append(np.mean(Q_scores))

results = pd.DataFrame({
    'k (truncation)': k_values,
    'Mean T²':        np.round(mean_T2, 4),
    'Mean Q (SPE)':   np.round(mean_Q, 4)
})

print("\nT² and Q Statistics vs Truncation Cutoff k:")
print(results.to_string(index=False))

print("\nINTERPRETATION:")
print("  k=1 → k=2: Mean Q drops sharply.")
print("    PC2 captures process B (Sensor 3). Moving from k=1 to k=2")
print("    removes a large structured residual from the Q space.")
print("    T² rises because the subspace now spans more real variance.")
print()
print("  k=2 → k=3: Mean Q flattens (the elbow).")
print("    PC3 onwards captures only noise. Adding more components")
print("    barely reduces Q because there is no more structure left.")
print("    This flat elbow at k=2 identifies the true physical")
print("    dimensionality of the system — two real latent processes.")
print()
print("  If k=3 is chosen incorrectly:")
print("    Pure noise (from Sensor 4 / Sensor 5) is absorbed into")
print("    the principal subspace. T² alarms become hypersensitive")
print("    to random fluctuations. Real faults in Q space are missed")
print("    because the residual subspace now contains less signal.")
````

As the truncation cutoff moves from k equal to 1 to k equal to 2, the Mean Q statistic drops sharply. This happens because the second principal component captures the second real latent process driving the plant, which in this case is the pressure loop reflected in Sensor 3. When only one component is retained, the entire contribution of this second process sits unaccounted for in the residual space, producing a large Q value. Adding the second component absorbs this structured variance, so the reconstruction error falls steeply. At the same time the Mean T squared rises because the subspace now spans more of the genuine process variation.
As the cutoff then moves from k equal to 2 to k equal to 3, the Mean Q statistic barely changes and levels off into a flat region. This is the elbow. The third component and everything beyond it captures only random sensor noise, which has no structured pattern to remove from the residual. Adding more components provides no meaningful reduction in reconstruction error because there is nothing real left to explain. The Q curve simply flattens because the subspace has already absorbed all the true physical information in the data.
This sharp drop followed by a flat elbow directly identifies k equal to 2 as the true hidden physical dimensionality of the asset. The drop tells you that real structured variance is being captured, and the plateau tells you that structure has been fully accounted for. Any further components are just fitting noise, not physics. The elbow is therefore the point where the principal subspace transitions from meaningful to meaningless.
If a truncation cutoff of k equal to 3 is incorrectly chosen, pure random noise is forced into the principal subspace. This has two damaging consequences. First, the T squared statistic becomes hypersensitive to normal random fluctuations, generating false alarms because the monitoring system now treats noise as part of the expected process behaviour. Second, the residual Q space is left with less genuine signal, making it harder to detect real process faults that should appear there. The monitoring system ends up simultaneously over-alarming on noise and under-alarming on actual faults, which is the worst possible outcome for a plant safety framework.

### Question 4: Operational Trade-offs in System Health Monitoring

* **Context:** Imagine you are deploying a real-time predictive maintenance pipeline for a fleet of identical industrial assets based on these exact models.
* **Core Problem:** Compare the operational trade-offs of these two approaches:
1. **The PCA Strategy:** Tracking the main principal subspace using Hotelling's $T^2$ alongside a residual noise floor monitor using the $Q$ statistic.
2. **The FA Strategy:** Monitoring the rotated Latent Factor Scores directly.


* Which strategy is more robust against a single sensor losing calibration or experiencing an electrical short? Justify your choice using the **Sensor Uniqueness Noise Floor ($\varphi^2$)** metrics.

````python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA, FactorAnalysis
from sklearn.preprocessing import StandardScaler

np.random.seed(42)
n_samples = 1000

# Two true latent plant processes
process_A = np.sin(np.linspace(0, 20, n_samples))
process_B = np.cos(np.linspace(0, 10, n_samples))

# Normal operating data
sensor_1 = 0.90 * process_A + np.random.normal(0, 0.20, n_samples)
sensor_2 = 0.85 * process_A + np.random.normal(0, 0.25, n_samples)
sensor_3 = 0.92 * process_B + np.random.normal(0, 0.20, n_samples)
sensor_4 = np.random.normal(0, 0.30, n_samples)  # noise sensor

df = pd.DataFrame({
    'Sensor_1': sensor_1, 'Sensor_2': sensor_2,
    'Sensor_3': sensor_3, 'Sensor_4': sensor_4
})

# Standardise
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

print("--- Operational Trade-offs: PCA Strategy vs FA Strategy ---\n")

# ── FA: Uniqueness (noise floor) per sensor ───────────────────────────────
fa = FactorAnalysis(n_components=2, rotation='varimax', random_state=42)
fa.fit(X_scaled)
uniqueness = fa.noise_variance_

print("SENSOR UNIQUENESS (ψ²) — Noise Floor per Sensor:")
for sensor, val in zip(df.columns, uniqueness):
    bar  = '█' * int(val * 20)
    flag = " ← high noise floor" if val > 0.50 else " ← structured sensor"
    print(f"  {sensor}: {val:.2%}  {bar}{flag}")

# ── Simulate a calibration fault on Sensor 2 ─────────────────────────────
# Sensor 2 loses calibration: its readings drift by a large offset
X_faulty          = X_scaled.copy()
X_faulty[:, 1]   += 4.0   # Sensor 2 drifts hard (electrical short / miscalibration)

# ── PCA Strategy: T² and Q scores ────────────────────────────────────────
pca = PCA(n_components=2)
pca.fit(X_scaled)

# Normal baseline scores
scores_normal     = pca.transform(X_scaled)
eigenvalues       = pca.explained_variance_
T2_normal         = np.mean(np.sum((scores_normal ** 2) / eigenvalues, axis=1))
recon_normal      = pca.inverse_transform(scores_normal)
Q_normal          = np.mean(np.sum((X_scaled - recon_normal) ** 2, axis=1))

# Faulty scores
scores_faulty     = pca.transform(X_faulty)
T2_faulty         = np.mean(np.sum((scores_faulty ** 2) / eigenvalues, axis=1))
recon_faulty      = pca.inverse_transform(scores_faulty)
Q_faulty          = np.mean(np.sum((X_faulty - recon_faulty) ** 2, axis=1))

print("\nPCA STRATEGY — T² and Q under normal vs sensor fault:")
print(f"  Mean T²  — Normal: {T2_normal:.3f}  |  Faulty: {T2_faulty:.3f}  "
      f"| Change: +{T2_faulty - T2_normal:.3f}")
print(f"  Mean Q   — Normal: {Q_normal:.3f}   |  Faulty: {Q_faulty:.3f}   "
      f"| Change: +{Q_faulty - Q_normal:.3f}")
print("  Sensor 2 loads on PC1 → fault contaminates the shared subspace.")
print("  Both T² and Q rise but it is unclear WHICH sensor caused it.")

# ── FA Strategy: Latent Factor Scores ────────────────────────────────────
fa_scores_normal  = fa.transform(X_scaled)
fa_scores_faulty  = fa.transform(X_faulty)

factor_normal_mean = np.mean(np.abs(fa_scores_normal), axis=0)
factor_faulty_mean = np.mean(np.abs(fa_scores_faulty), axis=0)

print("\nFA STRATEGY — Rotated Factor Scores under normal vs sensor fault:")
for i, (n, f) in enumerate(zip(factor_normal_mean, factor_faulty_mean)):
    print(f"  Factor {i+1} — Normal: {n:.3f}  |  Faulty: {f:.3f}  "
          f"| Change: +{f - n:.3f}")
print("  Sensor 2 loads exclusively on Factor 1 (ψ² is low → structured).")
print("  Factor 1 spikes cleanly. Factor 2 is unaffected.")
print("  Operator immediately knows: fault is in the Factor 1 sensor group.")

# ── Uniqueness as robustness metric ──────────────────────────────────────
print("\nROBUSTNESS COMPARISON USING UNIQUENESS (ψ²):")
print("  PCA strategy:")
print("    No uniqueness concept. Sensor 2's fault spreads across PC1 and PC2")
print("    because PCA mixes all sensors into every component.")
print("    Root cause isolation requires post-hoc contribution analysis.")
print()
print("  FA strategy:")
print("    Uniqueness ψ² tells us exactly how much each sensor's variance")
print("    is isolated from the shared factors.")
print("    Sensor 2 has low ψ² → it is tightly coupled to Factor 1.")
print("    A calibration fault on Sensor 2 appears as a clean Factor 1 spike.")
print("    Sensor 4 has high ψ² → its faults stay in its own noise floor")
print("    and do NOT contaminate the shared factor scores at all.")
print()
print("CONCLUSION:")
print("  FA strategy is more robust against single sensor faults.")
print("  Low ψ² sensors are tightly coupled to one factor → fault is isolated.")
print("  High ψ² sensors are decoupled → fault cannot pollute shared monitoring.")
print("  PCA has no equivalent mechanism — any sensor fault contaminates")
print("  all principal components simultaneously.")
````

The PCA strategy monitors the principal subspace using Hotelling's T squared statistic to track behaviour inside the subspace, and the Q statistic to track residual reconstruction error outside it. The FA strategy instead monitors the rotated latent factor scores directly, where each factor corresponds to a distinct physical subsystem of the plant.
When a single sensor loses calibration or experiences an electrical short, the two strategies respond very differently. In the PCA strategy, because every sensor loads partially on every principal component, a fault in one sensor spreads its effect across all components simultaneously. Both T squared and Q rise, but there is no clear signal pointing to which sensor or which subsystem caused the deviation. The operator must perform additional post-hoc contribution analysis to isolate the root cause, which costs time and requires statistical expertise during a live fault event.
In the FA strategy, the rotated factor structure means each sensor loads heavily on one factor and near-zero on all others. When Sensor 2 develops a calibration fault, it is tightly coupled to Factor 1 through a low uniqueness value, meaning its variance is almost entirely shared with that factor. The fault therefore appears as a clean, isolated spike on Factor 1 alone, while Factor 2 remains completely unaffected. The operator immediately knows the fault belongs to the Factor 1 sensor group and can act without any further analysis.
The sensor uniqueness noise floor makes this distinction precise. A sensor with low uniqueness is tightly coupled to a shared factor, so any fault in that sensor produces a strong and localised response in exactly one factor score. A sensor with high uniqueness, like Sensor 4, has most of its variance isolated in its own noise floor and completely decoupled from the shared factors. A fault or electrical short on a high uniqueness sensor therefore cannot contaminate the shared factor scores at all, because the monitoring system already treats that sensor's variance as independent noise.
The PCA strategy has no equivalent mechanism. It has no concept of uniqueness and no way to isolate one sensor's contribution from the shared subspace. Any sensor fault, regardless of which sensor or what type of failure, contaminates all principal components simultaneously and raises both T squared and Q in an ambiguous way.
The FA strategy is therefore more robust against single sensor faults. The uniqueness metric acts as a natural firewall. Structured sensors with low uniqueness produce clean, interpretable factor spikes that point directly to the affected subsystem. Noisy sensors with high uniqueness are quarantined from the shared monitoring space entirely. This combination gives FA-based monitoring a fault isolation capability that PCA fundamentally cannot match.
